```python
import pandas as pd
import matplotlib.pyplot as mp
import numpy as np
import opendatasets as od
import seaborn as sns
```

```python
od.download("https://www.kaggle.com/datasets/harikrishna9/covid19-dataset-by-john-hopkins-university")
```

```python
corona_df = pd.read_csv("covid19-dataset-by-john-hopkins-university/covid19_Confirmed_dataset.csv")
```

```python
corona_df.head()
```

```python
corona_df.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
corona_df
```

```python
corona=corona_df.groupby("Country/Region").sum()
corona.head(3)
```

```python
corona.loc["China"].plot(figsize=(4,4))
corona.loc["Italy"].plot()
corona.loc["Algeria"].plot()
mp.legend()
```

```python
corona.loc["China"][:3].plot(figsize=(3,3))
```

```python
corona.loc["China"].diff().plot(figsize=(3,3))
```

```python
corona.loc["China"].diff().max()
```

```python
corona.loc["Spain"].diff().max()
```

```python
corona.loc["Italy"].diff().max()
```

```python
countries= list(corona.index)
max_infection_rate= []

for c in countries:
    max_infection_rate.append(corona.loc[c].diff().max())

max_infection_rate
```

```python
corona["max_infection_rate"]= max_infection_rate
corona.head(3)
```

```python
corona_data= pd.DataFrame(corona["max_infection_rate"])
corona_data.head(3)
```

# Importing second Dataset

```python
world_happiness=pd.read_csv("covid19-dataset-by-john-hopkins-university/worldwide_happiness_report.csv")
world_happiness.head(3)
```

```python
happiness_data=world_happiness.drop(["Overall rank","Score","Generosity","Perceptions of corruption"],axis=1)
happiness_data.head(3)
```
